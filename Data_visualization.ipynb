```python
# Generate the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set up the output file (e.g., "sine_wave.html")
output_file("sine_wave.html")

# Create a Bokeh plot
p = figure(title="Sine Wave")

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.palettes import Inferno256 # Changed the import statement to import from bokeh.palettes

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create color mapper
mapper = LinearColorMapper(palette=Inferno256, low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Heatmap", x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           toolbar_location=None, width=400, height=400)

# Add heatmap
p.image(image=[data_heatmap], x=x.min(), y=y.min(), dw=x.max()-x.min(), dh=y.max()-y.min(),
        color_mapper=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Display in notebook
output_notebook()
show(p)
```
